```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Resultados manual o cargados
results = {
    "Modelo": ["Baseline CNN", "ResNet18"],
    "Accuracy": [0.5610, 0.5854],
    "Precision": [0.3147, 0.40],   
    "Recall": [0.5610, 0.5854],
    "F1-score": [0.4032, 0.48]    
}

df_results = pd.DataFrame(results)
df_results
```

```python
#Tabla comparativa
plt.figure(figsize=(6,2))
sns.heatmap(df_results.set_index("Modelo"), annot=True, cmap="Blues", fmt=".3f")
plt.title("Comparación de métricas: Baseline vs ResNet")
plt.show()
```

```python
#Comparación grafica
plt.figure(figsize=(8,5))
x = df_results["Modelo"]
plt.bar(x, df_results["Accuracy"], label="Accuracy", alpha=0.7)
plt.bar(x, df_results["F1-score"], label="F1-score", alpha=0.7)
plt.ylabel("Score")
plt.title("Comparación de desempeño")
plt.legend()
plt.show()
```
